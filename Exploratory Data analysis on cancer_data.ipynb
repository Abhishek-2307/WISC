## Cleaning the data

```python
df1 = pd.read_csv('cancer_data_means.csv')
# check whihc columns have missing values with info
df1.info()
```

```python
sum(df.duplicated())
```

```python
## remove "_mean" from columns names
new_labels = []
for col in df1.columns:
    if '_mean' in col:
        new_labels.append(col[:-5])  # exclude last 6 characters
    else:
        new_labels.append(col)
        
# new labels for our columns
new_labels
```

```python
# assign new labels to columns in dataframe
df1.columns = new_labels

# display first few rows of dataframe  to confirm changes
df.head()
```

```python
df1.to_csv('cancer_data_edit.csv', index=False)
```
